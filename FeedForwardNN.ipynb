## Question 1 : Download the Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

category_items = ['T-shirt', 'Trouser', 'Pullover-Shirt', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

item_sample_set = {}
num_classes = len(category_items)
for i in range(num_classes):
  item_sample_set[i] = np.where(y_train == i)[0][0]

for i in range(10):
  plt.subplot(2, 5, i+1)
  plt.imshow(X_train[item_sample_set[i]], cmap=plt.get_cmap('gray'))
  plt.title(category_items[i])

plt.show()
```
